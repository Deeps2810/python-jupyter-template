### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_data():
    """Generates x values and computes y values as x squared."""
    try:
        x = np.arange(0, 11)
        y = np.square(x)
        return x, y
    except Exception as e:
        print(f"Error generating data: {e}")
        return None, None

def plot_data(x, y):
    """Plots the data with error handling."""
    try:
        if x is None or y is None:
            raise ValueError("Invalid data. Cannot plot.")
        
        plt.figure(figsize=(8, 5))
        plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

        plt.title('Line Plot of y = x²')
        plt.xlabel('X Values')
        plt.ylabel('Y Values')
        plt.grid(True)
        plt.legend()
        plt.show()
    except Exception as e:
        print(f"Error plotting data: {e}")

def main():
    """Main function to generate and plot data."""
    x, y = generate_data()
    plot_data(x, y)

if __name__ == "__main__":
    main()
```
